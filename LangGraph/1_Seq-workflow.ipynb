# Sequencial Workflow (LangGraph)

```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
# define state first rule of any agent(we need to define state)

class BMIState(TypedDict):
    weight_kg: float
    height_m: float
    bmi_result: float
```

```python
# Node-1
def calculate_bmi(state: BMIState) -> BMIState:

    weight = state['weight_kg']
    height = state['height_m']

    bmi_result = weight / (height**2)

    state['bmi_result'] = round(bmi_result,2)

    return state
```

```python
# define Graph 
graph = StateGraph(BMIState) 

# 1-> add nodes to our graph
graph.add_node('calculate_bmi', calculate_bmi)

# 2-> add egdes between nodes
graph.add_edge(START, 'calculate_bmi')
graph.add_edge('calculate_bmi', END)
 
# 3-> compile the graph
workflow = graph.compile()
```

```python
# 4-> excute the graph
intial_state = {'weight_kg':58, 'height_m': 1.5}
final_state = workflow.invoke(intial_state)

print(final_state['bmi_result'])
```
